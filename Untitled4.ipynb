```python
# Import pandas for data manipulation
import pandas as pd

# Load the breast cancer dataset
print("[Loading Dataset]\n")
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/breast-cancer.csv')
print("[INFO] Dataset Loaded Successfully.\n")

# Display the first 5 rows of the dataset
print("[INFO] First 5 Rows of the Dataset:")
print(df.head())

# Display dataset structure and basic info
print("\n[INFO] Dataset Information:")
print(df.info())

# Display missing values in each column
print("\n[INFO] Missing Values per Column:")
print(df.isnull().sum())

# Summary statistics of the dataset
print("\n[INFO] Summary Statistics of the Dataset:")
print(df.describe())
```

```python
# Handle missing values by filling them with column means
print("\n[Handling Missing Values]\n")
numeric_df = df.select_dtypes(include=['number'])
df[numeric_df.columns] = numeric_df.fillna(numeric_df.mean())
print("[INFO] Missing values filled with column means.\n")

# Confirm no missing values remain
print("[INFO] Missing Values after Handling:")
print(df.isnull().sum())
```

```python
from sklearn.preprocessing import LabelEncoder

print("\n[Encoding Categorical Variables]\n")

# Label Encoding
label_encoders = {}
for col in df.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le
    print(f"[INFO] '{col}' column encoded using LabelEncoder.")

# One-Hot Encoding
print("\n[INFO] Applying One-Hot Encoding...")
df = pd.get_dummies(df, columns=df.select_dtypes(include=['object']).columns)
print("[INFO] One-Hot Encoding applied successfully.")
print("[INFO] Dataset after Encoding (First 5 Rows):")
print(df.head())
```

```python
print("\n[Splitting Dataset into Features and Target]\n")
target_column = df.columns[1]  # Assuming the second column is the target
X = df.drop(target_column, axis=1)
y = df[target_column]
print(f"[INFO] Target column: '{target_column}'")
print("[INFO] Sample of Features (First 5 Rows):")
print(X.head())
print("\n[INFO] Sample of Target (First 5 Rows):")
print(y.head())
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

print("\n[Scaling Features]\n")
standard_scaler = StandardScaler()
minmax_scaler = MinMaxScaler()

X_standard_scaled = standard_scaler.fit_transform(X)
X_minmax_scaled = minmax_scaler.fit_transform(X)

print("[INFO] StandardScaler and MinMaxScaler applied successfully.")
print("[INFO] First 5 Rows of Standard Scaled Features:")
print(pd.DataFrame(X_standard_scaled, columns=X.columns).head())
print("\n[INFO] First 5 Rows of MinMax Scaled Features:")
print(pd.DataFrame(X_minmax_scaled, columns=X.columns).head())
```

```python
from sklearn.model_selection import train_test_split

print("\n[Splitting Data into Training and Testing Sets]\n")
X_train, X_test, y_train, y_test = train_test_split(X_minmax_scaled, y, test_size=0.2, random_state=42)
print(f"[INFO] Training set size: {len(X_train)}, Test set size: {len(X_test)}")
print("[INFO] First 5 Rows of Training Features:")
print(pd.DataFrame(X_train, columns=X.columns).head())
print("\n[INFO] First 5 Rows of Training Target:")
print(y_train.head())
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
print("\n[Training and Evaluating Models]\n")

models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Decision Tree": DecisionTreeClassifier(),
    "Naive Bayes": GaussianNB()
}

performance_metrics = {"Model": [], "Accuracy": [], "Precision": [], "Recall": [], "F1 Score": []}

for model_name, model in models.items():
    print(f"\n[INFO] Training {model_name}...")
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    # Evaluate model performance
    accuracy = accuracy_score(y_test, y_pred)
    report = classification_report(y_test, y_pred, output_dict=True)

    print(f"[INFO] {model_name} Accuracy: {accuracy:.4f}")
    print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
    print("Classification Report:\n", classification_report(y_test, y_pred))

    # Store results
    performance_metrics["Model"].append(model_name)
    performance_metrics["Accuracy"].append(accuracy)
    performance_metrics["Precision"].append(report['weighted avg']['precision'])
    performance_metrics["Recall"].append(report['weighted avg']['recall'])
    performance_metrics["F1 Score"].append(report['weighted avg']['f1-score'])

    # Display Confusion Matrix
    disp = ConfusionMatrixDisplay(confusion_matrix=confusion_matrix(y_test, y_pred), display_labels=model.classes_)
    disp.plot(cmap='Blues', values_format='d')
    plt.title(f'Confusion Matrix for {model_name}')
    plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

print("\n[Visualizing Model Performance]\n")

metrics_df = pd.DataFrame(performance_metrics)
print("[INFO] Model Performance Metrics:")
print(metrics_df)

metrics_df.set_index('Model', inplace=True)
metrics_df.plot(kind='bar', figsize=(12, 8), colormap='viridis', width=0.8)
plt.title('Comparison of Model Performance')
plt.ylabel('Score')
plt.xlabel('Model')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
